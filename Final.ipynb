```python
# TODO: نمایش داده‌های Augmented
import math
def visualize_augmented_images(dataset, augmentation_layer, num_samples=9):

    images, labels = next(iter(dataset))
    batch_size = images.shape[0]
    n = min(num_samples, int(batch_size))
    aug = augmentation_layer(images, training=True)
    def to_uint8(x):
        x = tf.convert_to_tensor(x)
        xmax = tf.reduce_max(x)
        if tf.math.less_equal(xmax, 1.0):
            x = x * 255.0
        x = tf.clip_by_value(x, 0.0, 255.0)
        return tf.cast(x, tf.uint8)

    aug = to_uint8(aug)
    cols = int(math.ceil(math.sqrt(n)))
    rows = int(math.ceil(n / cols))
    plt.figure(figsize=(3*cols, 3*rows))
    for i in range(n):
        ax = plt.subplot(rows, cols, i + 1)
        plt.imshow(aug[i].numpy())
        plt.axis('off')
    plt.suptitle("Augmented Samples", y=0.98)
    plt.tight_layout()
    plt.show()
visualize_augmented_images(train_ds, data_augmentation, num_samples=9)
```

<div dir='rtl'>
  <font size=4>
    <li>
        این بخش از باید پیش‌پردازش (Preprocessing) روی داده‌های تصویری قبل از ورود به مدل انجام دهید.
         در اینجا هدف اصلی نرمال‌سازی مقادیر پیکسل‌ها است، به این صورت که مقادیر اصلی که در بازه [0, 255] هستند، به بازه [0.0, 1.0] تبدیل می‌شوند.
         یک تابع تعریف شده که هر تصویر را از نوع int به float تبدیل می‌کند و سپس آن را بر ۲۵۵ تقسیم می‌کند تا نرمال شود. این تابع را کامل کنید و سپس  روی کل داده‌های آموزش و اعتبارسنجی با استفاده از متد map اعمال شده است.
    </li>
  </font>
<br><br>

```python
# تعریف تابع پیش‌پردازش شامل نرمال‌سازی مقادیر پیکسل
def preprocess(image, label):
    image = tf.cast(image, tf.float32)
    image = image / 255.0
    return image, label

train_ds = train_ds.map(preprocess, num_parallel_calls=tf.data.AUTOTUNE)
val_ds   = val_ds.map(preprocess, num_parallel_calls=tf.data.AUTOTUNE)
```

```python
# TODO: ساخت مدل شبکه عصبی کانولوشنی (CNN) برای طبقه‌بندی تصاویر

NUM_CLASSES = len(class_names)
INPUT_SHAPE = IMG_SIZE + (3,)

data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.2),
    layers.RandomZoom(0.15),
    layers.RandomContrast(0.15),
    layers.RandomTranslation(0.1, 0.1),
], name="data_augmentation")
base = tf.keras.applications.MobileNetV2(
    input_shape=INPUT_SHAPE, include_top=False, weights="imagenet"
)
base.trainable = False
inputs = layers.Input(shape=INPUT_SHAPE)
x = data_augmentation(inputs)
x = layers.Lambda(lambda t: (tf.cast(t, tf.float32) * 2.0) - 1.0, name="to_minus1_1")(x)
x = base(x, training=False)
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dropout(0.4)(x)
outputs = layers.Dense(NUM_CLASSES, activation="softmax")(x)
model = models.Model(inputs, outputs, name="rps_mobilenetv2_tl")
```

```python
def plot_training_history(history):

    acc      = history.history['accuracy']
    val_acc  = history.history['val_accuracy']
    loss     = history.history['loss']
    val_loss = history.history['val_loss']
    epochs   = range(1, len(acc) + 1)
    plt.figure(figsize=(12,5))
    plt.subplot(1,2,1)
    plt.plot(epochs, acc, 'b-', label='Training Accuracy')
    plt.plot(epochs, val_acc, 'r-', label='Validation Accuracy')
    plt.title('Accuracy over Epochs')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True)
    plt.subplot(1,2,2)
    plt.plot(epochs, loss, 'b-', label='Training Loss')
    plt.plot(epochs, val_loss, 'r-', label='Validation Loss')
    plt.title('Loss over Epochs')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
plot_training_history(history)
```

```python
import random
def predict_random_samples(model, base_dir, class_names, k=2, img_size=(224,224), seed=None):
    if seed is not None:
        random.seed(seed)

    rows, cols = len(class_names), k
    plt.figure(figsize=(3*cols, 3*rows))

    for r, cls in enumerate(class_names):
        folder = os.path.join(base_dir, cls)
        files = [f for f in os.listdir(folder)
                 if f.lower().endswith((".jpg", ".jpeg", ".png", ".bmp", ".gif"))]

        if not files:
            ax = plt.subplot(rows, cols, r*cols + 1)
            ax.text(0.5, 0.5, f"No images in {cls}", ha="center", va="center")
            ax.axis("off")
            continue

        picks = random.sample(files, min(k, len(files)))

        for c, fname in enumerate(picks):
            path = os.path.join(folder, fname)

            img = tf.keras.utils.load_img(path, target_size=img_size)
            arr = tf.keras.utils.img_to_array(img) / 255.0
            batch = arr[None, ...]  # (1,H,W,C)

            probs = model.predict(batch, verbose=0)[0]
            idx = int(np.argmax(probs))
            pred = class_names[idx]
            conf = float(probs[idx]) * 100.0

            ax = plt.subplot(rows, cols, r*cols + c + 1)
            ax.imshow(img); ax.axis("off")
            ax.set_title(f"True: {cls}\nPred: {pred} ({conf:.1f}%)", fontsize=10)

    plt.suptitle(f"{k} random samples per class", y=0.995, fontsize=12)
    plt.tight_layout()
    plt.show()
predict_random_samples(model, "/content/drive/MyDrive/rps",
                             class_names, k=3, img_size=IMG_SIZE, seed=42)
```
